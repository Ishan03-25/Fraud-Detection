```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
data_frame_model = data_frame.drop(columns=['nameOrig', 'nameDest', 'isFlaggedFraud'], axis=1)
```

```python
categorical_features = ['type']
numerical_features = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
y = data_frame_model['isFraud']
X = data_frame_model.drop(columns=['isFraud'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numerical_features), ('cat', OneHotEncoder(drop='first'), categorical_features)], remainder='drop')
```

```python
model_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', LogisticRegression(class_weight='balanced', max_iter=1000))]) 
```

```python
model_pipeline.fit(X_train, y_train)
```

```python
y_predictions = model_pipeline.predict(X_test)
print(classification_report(y_test, y_predictions))
confusion_matrix(y_test, y_predictions)
model_pipeline.score(X_test, y_test)
```

```python
import joblib
joblib.dump(model_pipeline, 'fraud_detection_pipeline.pickle')
```
